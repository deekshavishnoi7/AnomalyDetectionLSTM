```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from keras.models import Sequential
from keras.layers import LSTM, RepeatVector, Dense,TimeDistributed, Input
from keras import regularizers
# from tensorflow.keras.layers import Input

import joblib
from sklearn.preprocessing import MinMaxScaler

# To read all files
data_dir = "Bearing_Sensor_Data/"
merged_data = pd.DataFrame()
```

```python
for filename in os.listdir(data_dir):
    dataset = pd.read_csv(os.path.join(data_dir, filename), sep='\t')
    dataset_abs_mean = np.array(dataset.abs().mean())
    dataset_abs_mean= pd.DataFrame(dataset_abs_mean.reshape(1,4))
    dataset_abs_mean.index = [filename]
    merged_data = merged_data.append(dataset_abs_mean)
```

```python
train = merged_data['2004-02-12 10:52:39': '2004-02-16 02:52:39']
test = merged_data['2004-02-16 02:52:39':]
print("Training dataset shape:", train.shape)
print("Test dataset shape:", test.shape)
```

```python
print(train.shape,test.shape)
train.plot(figsize = (20,10), title = "vibration")
```

```python
# scaled data using MinMaxScaler
```

```python
scaler = MinMaxScaler()
data_scaled =scaler.fit(train)
```

```python
train_scaled = data_scaled.transform(train)
test_scaled = data_scaled.transform(test)
```

```python
train_scaled.shape
```

```python
test_scaled.shape
```

```python
#  Reshape inputs for LSTM(sample, Time_stamp, Features)
train_reshape = train_scaled.reshape(train.shape[0], 1, train.shape[1])
print("Training shape:", train_reshape.shape)
test_reshape = test_scaled.reshape(test.shape[0], 1, test.shape[1])
print("Testing shape:", test_reshape.shape)
```

```python
# define the autoencoder network model
def autoencoder_model(X):
    inputs = Input(shape=(X.shape[1], X.shape[2]))
    L1 = LSTM(16, activation='relu', return_sequences=True, kernel_regularizer=regularizers.l2(0.00))(inputs)
    L2 = LSTM(8, activation='relu', return_sequences=True)(L1)
    L3 = LSTM(4, activation='relu', return_sequences=False)(L2)
    L4 = RepeatVector(X.shape[1])(L3)
    L5 = LSTM(4, activation='relu', return_sequences=True)(L4)
    L6 = LSTM(8, activation='relu', return_sequences=True)(L5)
    L7 = LSTM(16, activation='relu', return_sequences=True)(L6)
    output = TimeDistributed(Dense(X.shape[2]))(L7)    
    model = Model(inputs=inputs, outputs=output)
    return model
```

```python
# create the autoencoder model
model = autoencoder_model(train_reshape)
model.compile(optimizer='adam', loss='mae')
model.summary()
```

```python
history = model.fit(train_reshape,train_reshape, 
                    epochs=50, 
                    batch_size=20, 
                    validation_split=0.05).history
```

```python
# calculate loss of training data

train_pred = model.predict(train_reshape)
train_pred = train_pred.reshape(train_pred.shape[0], train_pred.shape[2])
train_pred = pd.DataFrame(train_pred, columns=train.columns)
train_pred.index = train.index
```

```python
# calculate loss of testing data

test_pred = model.predict(test_reshape)
test_pred = test_pred.reshape(test_pred.shape[0], test_pred.shape[2])
test_pred = pd.DataFrame(test_pred, columns=test.columns)
test_pred.index = test.index
```

```python
# plor Loss Distribution on the training data
anomaly_df1 = pd.DataFrame(train)
Xtrain = train_reshape.reshape(train_reshape.shape[0], train_reshape.shape[2])
anomaly_df1['Loss_mae'] = np.mean(np.abs(train_pred-Xtrain), axis = 1)
threshold = np.max(anomaly_df1['Loss_mae'])
print('Threshold :',threshold)
plt.figure(figsize=(16,9), dpi=80)
plt.title('Loss Distribution', fontsize=16)
sns.histplot(anomaly_df1['Loss_mae'], bins = 30, kde= True, color = 'blue')
plt.xlim(0, 0.5)
```

```python
threshold_min = np.std(anomaly_df1['Loss_mae'])
print('Threshold :',threshold_min)
```
